```python
Tracks = []

for i in range(50):
   Tracks.append(stephen['items'][i]['track']['name'])

TrackID = []

for i in range(50):
   TrackID.append(stephen['items'][i]['track']['id'])
```

```python
Artists = []

for i in range(50):
   Artists.append(stephen['items'][i]['track']['artists'][0]['name'])
```

```python
Albums = []


for i in range(50):
   Albums.append(stephen['items'][i]['track']['album']['name'])
```

```python
features = []
for i in range(0,len(TrackID),50):
    audio_features = spotify.audio_features(TrackID[i:i+50])
    for track in audio_features:
        features.append(track)
        features[-1]['target'] = 1
```

```python
audiofeat_df = pd.DataFrame(features)
audiofeat_df
```
